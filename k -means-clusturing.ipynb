```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

working_direc = os.getcwd()
file_name = "\\k-means.xlsx"
file_path = working_direc + file_name
print(file_path)
```

```python
df = pd.read_excel(file_path)
print(df)

plt.scatter(df['x'], df['y'])
plt.title('X-Y')
plt.xlabel('---x---')
plt.ylabel('---y---')
plt.grid(True)
plt.show()
```

### K-Means-Clusturing


```python
k = 3
model = KMeans(n_clusters = k).fit(df)
centroids = model.cluster_centers_
print(centroids)
```

```python
plt.scatter(df['x'], df['y'], c = model.labels_, s =100, alpha = 0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c = 'red', s = 50)
plt.grid()
plt.show()
```
